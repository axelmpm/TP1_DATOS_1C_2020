```python
from transformers import pipeline
import pandas as pd
```

```python
#SETTINGS
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
nlp = pipeline('sentiment-analysis')
```

```python
nlp('i am feeling great!')
```

```python
nlp('i am feeling sad :(')
```

```python
mapNeg = lambda x: 1 if nlp(x)[0]['label'] == 'NEGATIVE' else 0
```

```python
data = pd.read_csv('../../Data/train.csv')
```

```python
data = data.fillna(' ')
data['feature'] = data.apply(lambda x: x.keyword + ' ' + x.location + ' ' + x.text, axis = 1)
```

```python
data['prediction'] = data.feature.apply(lambda x: mapNeg(x))
```

```python
data.to_csv('sentiment')
```

```python
tp = data.apply(lambda x: 1 if (x.prediction == 1 and x.target == 1) else 0 ,axis = 1).sum()
fp = data.apply(lambda x: 1 if (x.prediction == 1 and x.target == 0) else 0 ,axis = 1).sum()
fn = data.apply(lambda x: 1 if (x.prediction == 0 and x.target == 1) else 0 ,axis = 1).sum()
```

```python
data[ data.prediction.eq(1) & data.target.eq(0) ]['text']
```

```python
precision = tp / (tp + fp)
```

```python
recall = tp / (tp + fn)
```

```python
f1 = 2 * precision * recall / (precision + recall)
```
