```python
import pandas as pd
```

# CARGO LOS FEATURES


```python
df_FEATURE1_train=pd.read_csv('train_final_featureGenerator1.csv')
df_FEATURE2_train=pd.read_csv('FE_v2_train.csv')

df_FEATURE1_toPredict=pd.read_csv('to_predict_featureGenerator1.csv')
df_FEATURE2_toPredict=pd.read_csv('FE_v2_test.csv')
df_FEATURE1_toPredict['id']=df_FEATURE2_toPredict.id
```

# MERGEO LOS FEATURES

```python
train=df_FEATURE2_train.merge(df_FEATURE1_train,on=['id','target'])*1
to_predict=df_FEATURE2_toPredict.merge(df_FEATURE1_toPredict,on='id')*1
```

# GUARDO LOS CSV PROCESADOS

```python
train.to_csv('train_feature_merged.csv',index=False)
to_predict.to_csv('to_predict_feature_merged.csv',index=False)
```
