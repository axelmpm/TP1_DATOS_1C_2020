```python
#CARGA DATA

A__C_tweets = pd.read_csv(generateAbsolutePath("train.csv"))
```

```python
#BUSCANDO LOS LOS USUARIOS MAS POPULARES

A_C_tweets1 = A__C_tweets.copy()

A_C_tweets1['appearing users'] = A_C_tweets1.apply(lambda row : extractUsers(row['text'],"@"), axis = 1)
A_C_tweets1 = A_C_tweets1[A_C_tweets1.apply(lambda row : len(row['appearing users']) > 0, axis = 1)]
A_C_tweets1 = A_C_tweets1['appearing users'].sum()

A_C_usersDF = getDataFrameOfFrecuencies('Usuarios',A_C_tweets1).rename(
                                    columns = {'Cantidad' : 'Cantidad de Menciones en Tweets'})
```

```python
A_C_TOP_AMOUNT = 10
A_C_topUsers = A_C_usersDF.head(A_C_TOP_AMOUNT)
```

```python
barplot('Cantidad de Menciones en Tweets','Usuarios',A_C_topUsers, title = 'Usuarios mas populares')
```
