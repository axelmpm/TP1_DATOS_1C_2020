# ANALISIS BASICO (Validez = 1 palabra en ingles)

''' ¿Que constituye una pregunta? Un tweet que tenga al menos 1 palabra en ingles antes de un signo de pregunta.'''
''' ¿Que constituye una afirmacion? Un tweet que tenga al menos 1 palabra en ingles.'''

```python
df = pd.read_csv(generateAbsolutePath("train.csv"))
df.text.isnull().sum() #no hay textos vacios
```

```python
df['textosValidos'] = df.text.apply(hasEnglishWord) 
df
```

```python
df['esPregunta'] = df.text.apply(isQuestion) 
```

```python
df['Afirmaciones'] = df.esPregunta * df.textosValidos #los que soon validos y no son preguntas, son afirmaciones
df
```

```python
df['preguntasConDesastre'] = df.esPregunta * df.target * df.textosValidos
df['preguntasSinDesastre'] = df.esPregunta * (df.target.apply(lambda x: 1 if x == 0 else 0)) * df.textosValidos
df['preguntasSinDesastre'].sum()
```
